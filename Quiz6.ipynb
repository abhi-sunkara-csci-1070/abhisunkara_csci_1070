```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("titanic.csv")
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

a = LabelEncoder()

data['Sex'] = a.fit_transform(data['Sex'].values)
data.head()
```

```python
b = data[['Survived', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex']].copy()
c = data[['Survived']]
b.corr()
```

```python
from statsmodels.formula.api import logit
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

f = data[['Sex', 'Pclass']]

y = np.array(data['Survived']).reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(f, y, test_size=0.3, random_state=42)
```

```python
regression = LogisticRegression(random_state=42).fit(X_train, y_train)
m = regression.predict(X_test)
```

I did logistic regression because it's discrete. I didnt do linear regression because it is continous. You cannot use a regression that is continous because this particualr code requires 2 specific outcomes.
