```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading a file function
def read_file(FILE_PATH):
    return pd.read_csv(FILE_PATH)

#Value count in function
def value_counting(arr):
    return arr.value_counts()
```

```python
data = read_file('CSCI_Fall_2024_survey.csv')
data.head()
```

```python
#data.rename(columns = {'What state/country were you born in?':'Birth Place'}, inplace = True)
data['What state/country were you born in?'] = np.where(data['What state/country were you born in?'].str.contains('St', case=False), "Missouri",
                               np.where(data['What state/country were you born in?'].str.contains('Saint', case=False), "Missouri",
                                        np.where(data['What state/country were you born in?'].str.contains('Missouri', case=False), "Missouri",
                                        np.where(data['What state/country were you born in?'].str.contains('Wis', case=False), "Wisconsin",
                                                 np.where(data['What state/country were you born in?'].str.contains('New', case=False), "New Jersey",
                                                          np.where(data['What state/country were you born in?'].str.contains('Kan', case=False), "Kansas",
                                                                   np.where(data['What state/country were you born in?'].str.contains('Okla', case=False), "Oklahoma",
                                                                            np.where(data['What state/country were you born in?'].str.contains('TEX', case=False), "Texas",
                                                                                     np.where(data['What state/country were you born in?'].str.contains('Mexico', case=False), "Mexico",
                                                                                              np.where(data['What state/country were you born in?'].str.contains('Fra', case=False), "Germany",
                                                                                                       np.where(data['What state/country were you born in?'].str.contains('India', case=False), "India",
                                                                                                                np.where(data['What state/country were you born in?'].str.contains('Georgia', case=False), "Georgia",
                                                                                                                         np.where(data['What state/country were you born in?'].str.contains('Republic', case=False), "China",
                                                                                                                                  np.where(data['What state/country were you born in?'].str.contains('Cameroon', case=False), "Cameroon",
                                                                                                                                          np.where(data['What state/country were you born in?'].str.contains('Kazakh', case=False), "Kazakhstah", "n/a")))))))))))))))
```

```python
data['How many pets and of which kind do you have?'] = np.where(data['How many pets and of which kind do you have?'].str.contains('dog', case=False) & data['How many pets and of which kind do you have?'].str.contains('Cat', case=False), "Dogs & Cats",
                                         np.where(data['How many pets and of which kind do you have?'].str.contains('dog', case=False), "Dogs",
                                                   np.where(data['How many pets and of which kind do you have?'].str.contains('cat', case=False), "Cats",
                                                          np.where(data['How many pets and of which kind do you have?'].str.contains('dog', case=False), "Dogs",
                                                                    np.where(data['How many pets and of which kind do you have?'].str.contains('schnauzers', case=False), "Dogs",
                                                                              np.where(data['How many pets and of which kind do you have?'].str.contains('bernedoodles', case=False), "Dogs",
                                                                                        np.where(data['How many pets and of which kind do you have?'].str.contains('Grumpy', case=False), "Other/Unknown",
                                                                                                  np.where(data['How many pets and of which kind do you have?'].str.contains('no', case=False) | data['How many pets and of which kind do you have?'].str.contains('0'), "None",
                                                                                                           np.where(data['How many pets and of which kind do you have?'].str.contains('NaN', case=False), "None",
                                                                                                                    np.where(data['How many pets and of which kind do you have?'].str.contains('1'), "Other/Unknown",
                                                                                                                                      np.where(data['How many pets and of which kind do you have?'].str.contains('dog', case=False), "Dogs",
                                                                                                                                               np.where(data['How many pets and of which kind do you have?'].str.contains('three', case=False), "Dogs",
                                                                                                                                                        np.where(data['How many pets and of which kind do you have?'].str.contains('not anymore', case=False), "None",
                                                                                                                                                                 np.where(data['How many pets and of which kind do you have?'].str.contains('I have two dogs and', case=False), "Dogs", "n/a"))))))))))))))
```

```python
data['What is your favorite type of food?'] = np.where(data['What is your favorite type of food?'].str.contains('italian', case=False), 'European cuisine',
        np.where(data['What is your favorite type of food?'].str.contains('Chinese', case=False), 'Asian cuisine',
                 np.where(data['What is your favorite type of food?'].str.contains('Japanese', case=False), 'Asian cuisine',
                          np.where(data['What is your favorite type of food?'].str.contains('Indian', case=False), 'Asian cuisine',
                                   np.where(data['What is your favorite type of food?'].str.contains('korean', case=False), 'Asian cuisine',
                                            np.where(data['What is your favorite type of food?'].str.contains('American', case=False), 'North American cuisine',
                                                     np.where(data['What is your favorite type of food?'].str.contains('Mexican', case=False), 'North American cuisine', 'n/a')))))))
                                                      
```

```python
data['What is your major?'] = np.where(data['What is your major?'].str.contains('computer science', case=False), 'Computer Science',
                                       np.where(data['What is your major?'].str.contains('data science', case=False), 'Data Science',
                                                np.where(data['What is your major?'].str.contains('biology', case=False), 'Other',
                                                         np.where(data['What is your major?'].str.contains('cs/ finance', case=False), 'Computer Science',
                                                                  np.where(data['What is your major?'].str.contains('CS/Business', case=False), 'Computer Science',
                                                                           np.where(data['What is your major?'].str.contains('economics', case=False), 'Other',
                                                                                    np.where(data['What is your major?'].str.contains('buisness economics', case=False), 'Computer Science',
                                                                                             np.where(data['What is your major?'].str.contains('TBD', case=False), 'Other',
                                                                                                      np.where(data['What is your major?'].str.contains('comp sci', case=False), 'Computer Science', 'n/a')))))))))
                                                                                                    
                                       
```

```python
data["How old are you?"] = data["How old are you?"].replace({17: "Minor",
                                                             18: "Adult",
                                                             19: "Adult",
                                                             20: "Adult",
                                                             21: "Adult"})
```

```python
b = value_counting(data['What is your favorite type of food?'])
plt.figure(figsize=(5,6))
b.plot(kind="bar")
plt.show()
```

```python
#The histogram shows students and their favorite type of food, simplified into cuisines. Through this, kids in our class prefer asian cuisine the most
```

```python
Comp_Sci = ((data["What is your major?"] == "Computer Science").sum()).item()
Data_Sci = ((data["What is your major?"] == "Data Science").sum()).item()
Other = ((data["What is your major?"] == "Other").sum()).item()
j = [Comp_Sci,Data_Sci,Other]
x = pd.DataFrame(j, index=['Computer Science', 'Data Science', 'Other'], columns=['Class Majors'])
```

```python
x.plot(kind = "pie", subplots=True, figsize = (9,9), colors = ("Red", "Green", "Yellow"))
```

```python
#The pie chart displays class majors in our computer scienece class. As seen, Computer Science is the most populous major among students in our class,
#while data science is second.
```

```python
adult = ((data["How old are you?"] == "Adult").sum()).item()
minor = ((data["How old are you?"] == "Minor").sum()).item()
Ages = [minor,adult]
```

```python
plt.style.use('_mpl-gallery')

x = 17 + np.arange(2)
y = Ages


fig, p = plt.subplots(figsize=(5,4))

p.stem(x, y)

p.set(xlim=(16, 19), xticks=np.arange(17,19),
       ylim=(0, 27), yticks=np.arange(0,27))
plt.xlabel('Age Number')
plt.ylabel('# of students with age')


plt.show()
```

```python
#The stem leaf plot displays data on the number of students who are 17 or 18+. Clearly more students in the class are 18 and older.
```
