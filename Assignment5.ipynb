```python
#lambda function
import time
time1 = time.perf_counter()

a["age_bin (lambda)"] = a["Age"].apply(lambda x: "Kid" if x <= 13
                                   else "Teenager" if x <= 18
                                   else "Adult" if x <= 50
                                   else "Senior Citizen")
time2 = time.perf_counter()
print(time2-time1)
a.head(50)                                  
```

```python
#np where function
import time
time1 = time.perf_counter()

a["age_bin (np where)"] = np.where(a["Age"] <= 13, "Kid",
                        np.where(a["Age"] <= 18, "Teenager",
                                 np.where(a["Age"] <= 50, "Adult",
                                          "Senior Citizen")))
time2 = time.perf_counter()
print(time2-time1)
a.head(50)
```

### Findings
The lambda function has a faster time in comparison to the np where function. When testing, the lambda function had about 0.004993624999769963 seconds while the np where function had about 0.005519416998140514 seconds. Both functions produced the same results and are

```python
fig, axes = plt.subplots(4, 1, figsize=(8, 12), sharex=True)

age = ['Kid', 'Teenager', 'Adult', 'Senior Citizen']
counter = 0

for group in age:
    axes[counter].hist(a[a['age_bin (lambda)'] == group] ['Fare'])
    axes[counter].set_title(f'Fare Distribution for {group}')
    counter += 1

plt.show()
```

### Findings
Fares differ by age_bin in the sense that most of the sub plot histograms are skewed right, making it clear that the titanic had passengers that paid more than the average person on the titanic. The average fare price was near 100 british pounds or lower. Additonally, there were more Adults on the titanic than their were minors or senior citizens aboard, as the adult sub histogram has a higher frequency of adults who attended the titanic (Between 500 to 600).
